```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


from pathlib import Path
```

```python
filepath = Path('../data/merged_weather_bike_data.csv')

# Read and display Csv data

weather_bike_df = pd.read_csv(filepath, encoding='utf-8', delimiter=',',)

weather_bike_df.head() 
```

```python

weather_bike_df['trip_length'] = pd.to_numeric(weather_bike_df['trip_length'], errors='coerce')

weather_bike_df.head().sum
```

```python
sampled_data = weather_bike_df.sample(n=1000, random_state=42)

sampled_data 
```

```python
# Filter data for the three bike types: electric_bike, classic_bike, and docked_bike
bike_types = ['electric_bike', 'classic_bike', 'docked_bike']
filtered_bikes_df = weather_bike_df[weather_bike_df['rideable_type'].isin(bike_types)]

# Display the filtered DataFrame
filtered_bikes_df
```

```python
# Ensure 'trip_length' is a timedelta in the filtered DataFrame
filtered_bikes_df['trip_length'] = pd.to_timedelta(filtered_bikes_df['trip_length'])

# Convert to total minutes
filtered_bikes_df['trip_length_minutes'] = filtered_bikes_df['trip_length'].dt.total_seconds() / 60

# Check the updated DataFrame
print(filtered_bikes_df[['trip_length', 'trip_length_minutes']].head())
```

```python
# Filter for Electric_bike and get the highest rides by trip length
electric_bike_rides = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'electric_bike']
highest_electric_bike_rides = electric_bike_rides.sort_values(by='trip_length_minutes', ascending=False)

# Filter for Classic_bike and get the highest rides by trip length
classic_bike_rides = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'classic_bike']
highest_classic_bike_rides = classic_bike_rides.sort_values(by='trip_length_minutes', ascending=False)

# Filter for Docked_bike and get the highest rides by trip length
docked_bike_rides = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'docked_bike']
highest_docked_bike_rides = docked_bike_rides.sort_values(by='trip_length_minutes', ascending=False)

# Display the results
print("Highest Electric Bike Rides:")
print(highest_electric_bike_rides[['ride_id', 'rideable_type', 'trip_length_minutes']])

print("\nHighest Classic Bike Rides:")
print(highest_classic_bike_rides[['ride_id', 'rideable_type', 'trip_length_minutes']])

print("\nHighest Docked Bike Rides:")
print(highest_docked_bike_rides[['ride_id', 'rideable_type', 'trip_length_minutes']])
```

```python

# Filter for Electric_bike and get the lowest 20 rides by trip length
electric_bike_rides = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'electric_bike']
lowest_electric_bike_rides = electric_bike_rides.sort_values(by='trip_length_minutes', ascending=True).head(20)

# Filter for Classic_bike and get the lowest 20 rides by trip length
classic_bike_rides = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'classic_bike']
lowest_classic_bike_rides = classic_bike_rides.sort_values(by='trip_length_minutes', ascending=True).head(20)

# Filter for Docked_bike and get the lowest 20 rides by trip length
docked_bike_rides = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'docked_bike']
lowest_docked_bike_rides = docked_bike_rides.sort_values(by='trip_length_minutes', ascending=True).head(20)

# Display the results
print("Lowest Electric Bike Rides:")
print(lowest_electric_bike_rides[['ride_id', 'rideable_type', 'trip_length_minutes']])

print("\nLowest Classic Bike Rides:")
print(lowest_classic_bike_rides[['ride_id', 'rideable_type', 'trip_length_minutes']])

print("\nLowest Docked Bike Rides:")
print(lowest_docked_bike_rides[['ride_id', 'rideable_type', 'trip_length_minutes']])
```

```python
# Calculate summary statistics for 'trip_length_minutes' by 'rideable_type'
summary_stats = filtered_bikes_df.groupby('rideable_type')['trip_length_minutes'].describe()

# Print the summary statistics
print(summary_stats)


# Convert trip_length_minutes to days
filtered_bikes_df['trip_length_days'] = filtered_bikes_df['trip_length_minutes'] / 1440

# Calculate summary statistics for 'trip_length_days' by 'rideable_type'
summary_stats = filtered_bikes_df.groupby('rideable_type')['trip_length_days'].describe()

# Print the summary statistics
print(summary_stats)


```

```python
# Filter the DataFrame for docked bikes only
docked_bikes_df = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'docked_bike'].copy()

# Convert 'trip_length_minutes' to 'trip_length_days' using .loc to avoid SettingWithCopyWarning
docked_bikes_df.loc[:, 'trip_length_days'] = docked_bikes_df['trip_length_minutes'] / 1440  # Convert minutes to days

# Visualization: Box plot for docked bikes showing trip duration in days
plt.figure(figsize=(12, 6))
plt.boxplot(
    docked_bikes_df['trip_length_days'],
    patch_artist=True,
    boxprops=dict(facecolor='lightblue', color='darkblue'),
    medianprops=dict(color='red'),
    whiskerprops=dict(color='blue'),
    capprops=dict(color='blue')
)

# Add labels and title
plt.title('Trip Duration for Docked Bikes (in Days)', fontsize=16)
plt.xlabel('Docked Bikes', fontsize=12)
plt.ylabel('Trip Length (Days)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()

# Optionally, print the summary statistics for docked bikes
summary_stats_docked = docked_bikes_df['trip_length_days'].describe()
print(summary_stats_docked)
```

```python
# Filter the DataFrame for electric bikes only
electric_bikes_df = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'electric_bike'].copy()

# Filter the DataFrame for classic bikes only
classic_bikes_df = filtered_bikes_df[filtered_bikes_df['rideable_type'] == 'classic_bike'].copy()

# Visualization: Box plot for electric bikes showing trip duration in minutes
plt.figure(figsize=(12, 6))

# Electric Bike Boxplot
plt.subplot(1, 2, 1)
plt.boxplot(
    electric_bikes_df['trip_length_minutes'],
    patch_artist=True,
    boxprops=dict(facecolor='lightgreen', color='darkgreen'),
    medianprops=dict(color='red'),
    whiskerprops=dict(color='blue'),
    capprops=dict(color='blue')
)
plt.title('Trip Duration for Electric Bikes (in Minutes)', fontsize=14)
plt.ylabel('Trip Length (Minutes)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Classic Bike Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(
    classic_bikes_df['trip_length_minutes'],
    patch_artist=True,
    boxprops=dict(facecolor='lightcoral', color='darkred'),
    medianprops=dict(color='red'),
    whiskerprops=dict(color='blue'),
    capprops=dict(color='blue')
)
plt.title('Trip Duration for Classic Bikes (in Minutes)', fontsize=14)
plt.ylabel('Trip Length (Minutes)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Layout adjustments
plt.tight_layout()
plt.show()

# Optionally, print the summary statistics for electric and classic bikes
summary_stats_electric = electric_bikes_df['trip_length_minutes'].describe()
print("Electric Bike Summary Stats:")
print(summary_stats_electric)

summary_stats_classic = classic_bikes_df['trip_length_minutes'].describe()
print("Classic Bike Summary Stats:")
print(summary_stats_classic)

```

```python
# Group by start_hour, member_casual, and rideable_type to get ride counts
hourly_trends = weather_bike_df.groupby(['start_hour', 'member_casual', 'rideable_type']).size().reset_index(name='ride_count')

# Pivot data to have start_hour as index and columns for member_casual and rideable_type
pivot_data = hourly_trends.pivot_table(index='start_hour', columns=['member_casual', 'rideable_type'], values='ride_count', aggfunc='sum').fillna(0)

# Plotting the line chart
plt.figure(figsize=(14, 8))

# Plot trends for each rider type and rideable type
for rider_type in ['member', 'casual']:
    for rideable in weather_bike_df['rideable_type'].unique():
        column_name = (rider_type, rideable)
        
        if column_name in pivot_data.columns:
            plt.plot(pivot_data.index, pivot_data[column_name], label=f'{rider_type} - {rideable}', marker='o')

# Add labels, title, and legend
plt.title('Hourly Ride Trends by Rider Type and Bike Type', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Rides', fontsize=12)
plt.xticks(range(0, 24))  # Assuming start_hour is in 24-hour format
plt.legend(title='Rider Type and Bike Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot with tight layout
plt.tight_layout()
plt.show()
```
