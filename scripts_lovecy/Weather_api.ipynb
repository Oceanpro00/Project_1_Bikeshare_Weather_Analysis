```python
# Dependencies
import pandas as pd
import json
import requests
from API_keys import api_key
from API_keys import geoapify_key
from pprint import pprint
import hvplot.pandas
```

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
city = "Chicago"
units ="metric"

# Build query URL
query_url = f"{url}appid={api_key}&units={units}&q={city}"
```

```python
# Get weather data
weather_response = requests.get(query_url)
weather_json = weather_response.json()

# Print the response
pprint(f"The weather API responded with: {weather_json}.")
```

```python
weather_data = {
    "city": weather_json['name'],
    "lat": weather_json['coord']['lat'],
    "lon": weather_json['coord']['lon'],
    "Temperature" : weather_json['main']['temp'],
    "Humidity": weather_json['main']['humidity'],
    "Pressure" : weather_json['main']['pressure'],
    "Windspeed": weather_json['wind']['speed']
    }
chicago_data = pd.DataFrame([weather_data])
chicago_data

```

```python

latitude = 41.85
longitude = -87.65
radius = 5000

# Define Chicago's central coordinates and radius
params = {
    "categories" : "rental.bicycle",
    "filter" : f"circle:{longitude},{latitude},{radius}",
    "api_key": geoapify_key

}

# Geoapify Places API endpoint
base_url = f"https://api.geoapify.com/v2/places"

# Make the API request
response = requests.get(base_url,params=params)
data = response.json()
pprint(data)
```

```python
# Parse the response to extract docking station information
stations = []
for feature in data['features']:
    station = {
        "brand": feature['properties'].get('brand','unknown'),
        'name': feature['properties'].get('name', 'Unknown'),
        'lat': feature['geometry']['coordinates'][1],
        'lon': feature['geometry']['coordinates'][0]
    }
    stations.append(station)

# Convert the data to a DataFrame
stations_df = pd.DataFrame(stations)

# Display the DataFrame
stations_df
```

```python
Divvy_station = stations_df.loc[stations_df["brand"] == 'Divvy']
Divvy_station
```

```python



station_map = Divvy_station.hvplot.points(
    x="lon",           
    y="lat",             
    geo=True,               
    tiles="OSM",            
    color="name",          
    width = 1000,
    height = 600,
    alpha=0.6,              
    hover_cols=["latitude", "longitude", "name"] ,
)

# Display the map
station_map


```
